```python
import os
!pip show opencv-python

import cv2
print(cv2.__version__)

import matplotlib.pyplot as plt
import numpy as np
import random

# Set your data path
data_path = '../data/train'

# 1. List all emotion folders (directories only)
emotions = [e for e in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, e))]
print(f"Emotions: {emotions}")

# 2. Count images in each emotion
for emotion in emotions:
    emotion_path = os.path.join(data_path, emotion)
    count = len(os.listdir(emotion_path))
    print(f"{emotion}: {count} images")

# 3. Display sample images from happy folder
happy_path = os.path.join(data_path, 'happy')
happy_images = os.listdir(happy_path)
random_imgs = random.sample(happy_images, 3)

plt.figure(figsize=(15, 5))

for i, img_name in enumerate(random_imgs):
    pt = os.path.join(happy_path, img_name)   # FIXED PATH
    img = cv2.imread(pt)

    if img is None:
        print(f"Failed to load {img_name}")
        continue
    print(img.shape)

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(1, 3, i+1)
    plt.imshow(img_rgb)
    plt.title(f"Happy {i+1}")
    plt.axis('off')

plt.tight_layout()
plt.show()
# The dataset is severely imbalanced. The largest class has 7,215 samples while the smallest has only 436, giving a ratio of about 16.5. 
# This means the model will be biased toward majority classes and accuracy alone will be misleading. 
# Therefore, class weighting or other balancing techniques are necessary.
```
